## Regex

### Modo Verbose:
- **`(?x)`**: Esto habilita el modo verbose, que permite incluir comentarios y espacios en blanco en la expresión regular para mejorar la legibilidad.

### URLs:
- **`http[s]*?:\/\/\S+`**:
  - **`http[s]*?`**: Coincide con `http` o `https`.
  - **`:\/\/`**: Coincide con `://`, que es parte de la estructura de una URL.
  - **`\S+`**: Uno o más caracteres que no son espacios en blanco, representando el resto de la URL.
  - URLs que comienzan con `http://` o `https://`.

- **`\[http[^\]]+\]`**:
  - **`\[`** y **`\]`**: Coinciden con los corchetes literales.
  - **`http[^\]]+`**: URL que comience con "http" y termine con un corchete de cierre (`]`), sin incluir `]` dentro de la URL.
  - URLs que están entre corchetes, por ejemplo, `[http://example.com]`.

### Etiquetas
- **`<.*?>`**:
  - **`<`** y **`>`**
  - **`.*?`**: Coincide con cualquier carácter con la mínima longitud coincidente (lazy)
  - Esta parte captura tags usados para la tokenización

### Acrónimos:
- **`(?:[A-Z]\.)+`**:
  - **`(?: ... )`**: Un grupo no capturante. Agrupa la expresión sin capturarla.
  - **`[A-Z]`**: Letras mayúsculas
  - **`\.`**: Punto literal (`.`).
  - **`+`**: Repite el patrón una o mas veces

### Números (moneda, decimales, porcentajes):
- **`\$?\d+(?:\.\d+)?%?`**:
  - **`\$?`**: Coincide con un signo de dólar opcional (`$`).
  - **`\d+`**: Coincide con uno o más dígitos (un número entero).
  - **`(?:\.\d+)?`**: Un grupo no capturante que opcionalmente coincide con un punto decimal seguido de uno o más dígitos (decimales).
  - **`%?`**: Coincide con un signo de porcentaje opcional (`%`).

### Palabras con Guiones:
- **`\w+(?:-\w+)*`**:
  - **`\w+`**: Coincide con uno o más caracteres de palabra
  - **`(?:-\w+)*`**: Un grupo no capturante que coincide con un guion (`-`) seguido de más caracteres de palabra, y puede repetirse cero o más veces.

### Elipsis (`...`):
- **`\.\.\.`**:

```python
import re
pattern = r'''(?x)                     # Activar verbose mode
            (?:[A-Z]\.)+             # Abreviaturas como U.S.A.
            | \$?\d+(?:\.\d+)?%?      # Números con decimales opcionales y % (el % no es opcional aquí)
            | \w+(?:-\w+)*           # Palabras con guiones internos opcionales
            | \.\.\.                 # Elipsis (...)
            '''
text =  """
        Dr. John visited the U.S.A. in 2020. He spent $12.50 on food, 15% on tips, and used ellipses like this... a lot!
        https://colab.research.google.com/drive/1I-lbHN8iH8U10AZy_Vm9Bd76iLpIHr1l#scrollTo=50KINHoNNMx9
        """

urls = re.findall(r'http[s]*?:\/\/\S+|\[http[^\]]+\]', text)
text = re.sub(r'http[s]*?:\/\/\S+|\[http[^\]]+\]', "", text)    #URLs
#print(text)
text = re.sub(r'\.\.\.', "", text)                    # elipsis
#print(text)
text = re.sub(r'[\]\[,;\"\'\?\(\)\:\-_`]', "", text)      # tokens separados; incluye ], [
#print(text)

# Apply the pattern using re.findall to extract all matches
tokens = re.findall(pattern, text)
tokens = tokens + urls
print(tokens)
# Print the matched tokens
for token in tokens:
    print(token)
```

```python
import re

def re_tokenize(text):
    pattern = r'''(?x)                     # Activar verbose mode
              | <.*?>
              | \w+(?:-\w+)*           # Palabras con guiones internos opcionales
              '''

    #urls = re.findall(r'http[s]*?:\/\/\S+|\[http[^\]]+\]', text)
    text = re.sub(r'http[s]*?:\/\/\S+|\[http[^\]]+\]', "<URL>", text)    #URLs
    text = re.sub(r'(?:[A-Z]\.)+', "<ACRONYM>", text)
    #print(text)
    text = re.sub(r'\.\.\.', "", text)                    # elipsis
    #print(text)
    text = re.sub(r'\$?\d+(?:\.\d+)?%?', '<NUM>', text)      # Números con decimales opcionales y %
    #print(text)
    text = re.sub(r'[:;=Xx]-?[)vVdDpPOo\(\\|\/p]|\^_\^|<3|:\'\(|:-?[\\\/]|:-?[\|]\b', "<EMOTICON>", text)  # Emoticones

    text = re.sub(r'[\]\[\,\;\"\'\?\(\)\:\-\_\`]', "", text)      # tokens separados; incluye ], [
    #print(text)

    # Apply the pattern using re.findall to extract all matches
    tokens = re.findall(pattern, text)
    tokens = [token for token in tokens if token != '' ]

    #print(tokens)
    return tokens
```

## BPE

```python
from collections import defaultdict, Counter

class BytePairEncoding:
    def __init__(self, num_merges):
        self.num_merges = num_merges
        self.vocab = {}
        self.merges = []

    def train(self, corpus):
        # Creación inicial del vocabulario
        self.vocab = {}
        for word in corpus:
            word_with_space = " ".join(list(word))
            if word_with_space in self.vocab:
                self.vocab[word_with_space] += 1
            else:
                self.vocab[word_with_space] = 1

        print(f"Vocabulario inicial: {self.vocab}")

        for _ in range(self.num_merges):
            # Encuentra los pares de simbolos más  frecuentes

            pairs = self.get_stats(self.vocab)
            if not pairs:
                break

            best_pair = max(pairs, key=pairs.get)
            self.merges.append(best_pair)

            print(f"Par más frecuente: {best_pair}")

            # Fusionar el par más frecuente
            self.vocab = self.merge_vocab(best_pair, self.vocab)

            #print(f"Vocabulario actualizado: {self.vocab}")

    def get_stats(self, vocab):
        pairs = defaultdict(int)
        for word, freq in vocab.items():
            symbols = word.split()
            for i in range(len(symbols) - 1):
                pair = (symbols[i], symbols[i + 1])
                pairs[pair] += freq
        return pairs

    def merge_vocab(self, pair, vocab):
        bigram = re.escape(' '.join(pair))
        p = re.compile(r'(?<!\S)' + bigram + r'(?!\S)')
        new_vocab = {}
        for word in vocab:
            new_word = p.sub(''.join(pair), word)
            new_vocab[new_word] = vocab[word]
        return new_vocab

    def tokenize(self, word):
        if re.match(r'<.*?>', word):
            return [word]
        word = " ".join(list(word))
        symbols = word.split()

        for merge in self.merges:
            i = 0
            while i < len(symbols) - 1:
                if symbols[i] == merge[0] and symbols[i + 1] == merge[1]:
                    symbols[i:i + 2] = [''.join(merge)]
                i += 1

        return symbols

# Ejemplo de uso
corpus = ["hello", "hell", "helium", "hero", "her"]
num_merges = 3  # Número de combinaciones deseadas

# Creación de una instancia de BytePairEncoding
bpe = BytePairEncoding(num_merges)

bpe.train(corpus)
vocab = bpe.vocab

print("\nVocabulario final generado:")
print(vocab)

# Tokenización de una palabra nueva
word_to_tokenize = "helium"
tokens = bpe.tokenize(word_to_tokenize)
print(f"\nTokenización de '{word_to_tokenize}': {tokens}")
```

## Tokenizer

```python
class Tokenizer:
  def __init__(self):
    self.bpe = BytePairEncoding(200)


  def train(self, corpus):
    words = re_tokenize(corpus)
    words = [word for word in words if not re.match(r'<.*?>', word)]
    self.bpe.train(words)
    print(self.bpe.vocab)

  def tokenize(self, text):
    text = re_tokenize(text)
    print(text)
    full_tokens = []
    for word in text:
      tokens = self.bpe.tokenize(word)
      print(f"\nTokenización de '{word}': {tokens}")
      full_tokens = full_tokens + tokens
    return full_tokens
```

```python
tokenizer = Tokenizer()
corpus = ""
for text in wikidata["train"][:100]["text"]:
  corpus = corpus + text
tokenizer.train(corpus)
```
