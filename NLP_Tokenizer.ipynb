```python
import re
pattern = r'''(?x)                     # Activar verbose mode
            (?:[A-Z]\.)+             # Abreviaturas como U.S.A.
            | \$?\d+(?:\.\d+)?%?      # Números con decimales opcionales y % (el % no es opcional aquí)
            | \w+(?:-\w+)*           # Palabras con guiones internos opcionales
            | \.\.\.                 # Elipsis (...)

            '''
text =  """
        Dr. John visited the U.S.A. in 2020. He spent $12.50 on food, 15% on tips, and used ellipses like this... a lot!
        https://colab.research.google.com/drive/1I-lbHN8iH8U10AZy_Vm9Bd76iLpIHr1l#scrollTo=50KINHoNNMx9
        """

urls = re.findall(r'http[s]*?:\/\/\S+|\[http[^\]]+\]', text)
text = re.sub(r'http[s]*?:\/\/\S+|\[http[^\]]+\]', "", text)    #URLs
#print(text)
text = re.sub(r'\.\.\.', "", text)                    # elipsis

#print(text)
text = re.sub(r'[\]\[,;\"\'\?\(\)\:\-_`]', "", text)      # tokens separados; incluye ], [
#print(text)

# Apply the pattern using re.findall to extract all matches
tokens = re.findall(pattern, text)
tokens = tokens + urls
print(tokens)
# Print the matched tokens
for token in tokens:
    print(token)
```

```python
import re
def split_sentence_around_url(pattern, sentence):
    match = re.search(pattern, sentence)

    if match:
        url = match.group(0)
        before_url = sentence[:match.start()].strip()
        after_url = sentence[match.end():].strip()
        return before_url, url, after_url
    else:
        return sentence, None, None

def re_tokenize(text):
    pattern = r'''(?x)                     # Activar verbose mode
              http[s]*?:\/\/\S+|\[http[^\]]+\]
              | <.*?>

              | (?:[A-Z]\.)+             # Abreviaturas como U.S.A.
              | \$?\d+(?:\.\d+)?%?      # Números con decimales opcionales y % (el % no es opcional aquí)
              | \w+(?:-\w+)*           # Palabras con guiones internos opcionales

              | \.\.\.                 # Elipsis (...)
              '''

    #urls = re.findall(r'http[s]*?:\/\/\S+|\[http[^\]]+\]', text)
    text = re.sub(r'http[s]*?:\/\/\S+|\[http[^\]]+\]', "<URL>", text)    #URLs
    text = re.sub(r'(?:[A-Z]\.)+', "<ACRONYM>", text)
    #print(text)
    text = re.sub(r'\.\.\.', "", text)                    # elipsis
    #print(text)
    text = re.sub(r'\$?\d+(?:\.\d+)?%?', '<NUM>', text)      # Números con decimales opcionales y % (el % no es opcional aquí)
    #print(text)
    text = re.sub(r'[:;=Xx]-?[)vVdDpPOo\(\\|\/p]|\^_\^|<3|:\'\(|:-?[\\\/]|:-?[\|]\b', "<EMOTICON>", text)                    # elipsis
    text = re.sub(r'[\]\[\,\;\"\'\?\(\)\:\-\_\`]', "", text)      # tokens separados; incluye ], [
    #print(text)

    # Apply the pattern using re.findall to extract all matches
    tokens = re.findall(pattern, text)
    tokens = [token for token in tokens if token != '' ]

    #print(tokens)
    return tokens
```

## BPE

```python
from collections import defaultdict, Counter

class BytePairEncoding:
    def __init__(self, num_merges):
        self.num_merges = num_merges
        self.vocab = {}
        self.merges = []

    def train(self, corpus):
        """
        Train the BPE model on a corpus.
        :param corpus: List of words or sentences in the corpus.
        """
        # Create the initial vocabulary with all characters
        self.vocab = {}
        for word in corpus:
            word_with_space = " ".join(list(word))
            if word_with_space in self.vocab:
                self.vocab[word_with_space] += 1
            else:
                self.vocab[word_with_space] = 1

        #print(f"Initial Vocabulary: {self.vocab}")

        # Perform BPE merges
        for _ in range(self.num_merges):
            # Find the most frequent pair of symbols
            pairs = self.get_stats(self.vocab)
            if not pairs:
                break

            best_pair = max(pairs, key=pairs.get)
            self.merges.append(best_pair)

            #print(f"Most frequent pair: {best_pair}")

            # Merge the most frequent pair
            self.vocab = self.merge_vocab(best_pair, self.vocab)

            #print(f"Updated Vocabulary: {self.vocab}")

    def get_stats(self, vocab):
        """
        Get the frequency of pairs of symbols.
        :param vocab: Current vocabulary dictionary with word frequencies.
        :return: Dictionary of pairs with their frequencies.
        """
        pairs = defaultdict(int)
        for word, freq in vocab.items():
            symbols = word.split()
            for i in range(len(symbols) - 1):
                pair = (symbols[i], symbols[i + 1])
                pairs[pair] += freq
        return pairs

    def merge_vocab(self, pair, vocab):
        """
        Merge the most frequent pair in the vocabulary.
        :param pair: Pair to merge.
        :param vocab: Current vocabulary dictionary.
        :return: Updated vocabulary after the merge.
        """
        bigram = re.escape(' '.join(pair))
        p = re.compile(r'(?<!\S)' + bigram + r'(?!\S)')
        new_vocab = {}
        for word in vocab:
            new_word = p.sub(''.join(pair), word)
            new_vocab[new_word] = vocab[word]
        return new_vocab

    def tokenize(self, word):
        """
        Tokenize a word using the learned BPE merges.
        :param word: Word to tokenize.
        :return: List of BPE tokens.
        """
        if re.match(r'<.*?>', word):
            return [word]
        word = " ".join(list(word))
        symbols = word.split()

        for merge in self.merges:
            i = 0
            while i < len(symbols) - 1:
                if symbols[i] == merge[0] and symbols[i + 1] == merge[1]:
                    symbols[i:i + 2] = [''.join(merge)]
                i += 1

        return symbols

# Ejemplo de uso
corpus = ["low", "lowest", "new", "wider"]
num_merges = 4  # Número de combinaciones deseadas

bpe = BytePairEncoding(num_merges)
bpe.train(corpus)
vocab = bpe.vocab

print("\nVocabulario final generado:")
print(vocab)

# Tokenización de una palabra nueva
word_to_tokenize = "lowest"
tokens = bpe.tokenize(word_to_tokenize)
print(f"\nTokenización de '{word_to_tokenize}': {tokens}")
```

## Tokenizer

```python
class Tokenizer:
  def __init__(self):
    self.bpe = BytePairEncoding(200)


  def train(self, corpus):
    words = re_tokenize(corpus)
    words = [word for word in words if not re.match(r'<.*?>', word)]
    self.bpe.train(words)
    print(self.bpe.vocab)

  def tokenize(self, text):
    text = re_tokenize(text)
    print(text)
    full_tokens = []
    for word in text:
      tokens = self.bpe.tokenize(word)
      print(f"\nTokenización de '{word}': {tokens}")
      full_tokens = full_tokens + tokens
    return full_tokens


```
